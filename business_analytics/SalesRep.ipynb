```python
# load excel workbook and create new sheet
from openpyxl import load_workbook
wb2 = load_workbook('Sales.xlsx')
```

```python
# sheetnames
wb2.sheetnames
```

```python
# data size in Source sheet
print("Excel cell range: {}".format(wb2["Source"].dimensions))
print("Maximum row: {}".format(wb2["Source"].max_row))
print("Maximum column: {}".format(wb2["Source"].max_column))
```

```python
# dataframe to analyze
import pandas as pd
df = pd.DataFrame(wb2["Source"].values)
df.columns = df.iloc[0,:]
df = df.iloc[1:,:]
df.head()
```

```python
import qgrid

qgrid.show_grid(df, show_toolbar=True)
```

```python
# total revenue of each salesrep
group_name = df[["SalesRep","Revenue"]].groupby('SalesRep').sum()
group_name
```

```python
revenue = []
for i in group_name.values:
    revenue.append(int(i))
```
